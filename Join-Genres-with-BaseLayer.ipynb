```python
import pandas as pd
base = "base-layers/american-left-ephemera/item-base-layer.csv"
df_base = pd.read_csv(base, index_col="Unnamed: 0").fillna("")
df_base
```

```python
extension_raw = "extension-layers/american-left-ephemera/tagged_genres/genre_tags_raw.csv"
df_ext = pd.read_csv(extension_raw).fillna("")
df_ext
```

```python
df_join = df_ext[['assignee', 'link', 'genre', 'genre 2', 'level_of_confidence', 'identifier']].join(df_base.set_index('identifier'), on='identifier', lsuffix='_extended', rsuffix='')
```

```python
df_join = df_join.rename(columns={'genre 2':'genre_2_extended'})
df_join.to_csv("extension-layers/american-left-ephemera/tagged_genres/genre_tags_joined.csv")
```
